# Analyse des mots




regarde la fréquence d'apparition des mots dans un "spam" dont la longueur est plus que X

```python
import re

x = 3

dict_mot_spam = dict()

for i in range (default.size // 4):
    
    #check if spam
    if default.iloc[i][3] == 1:
        #retire les points, virgules, et autres caracteres de mise en page
        mot = re.findall(r"[\w]+",default.iloc[i][2])
        for w in mot:
            if len(w) > x:
                value = dict_mot_spam.get(w, None)
                if value is not None: 
                    dict_mot_spam[w] = value+1
                else:
                    dict_mot_spam[w] = 1
                
            
            
```

```python
print(dict_mot_spam)
```
