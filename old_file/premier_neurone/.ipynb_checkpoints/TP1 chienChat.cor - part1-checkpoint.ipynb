##  Perceptron simple à 1 neurone

1) Par minimisation de l'erreur moyenne quadratique

###  Propagation avant

```python
# On initialise les paramètres du perceptron
def initParam(n):
    W=np.random.randn(n,1)
    b=np.random.randn(1)
    return (W,b)
```

```python
W,b = initParam(Xr.shape[1])
print(W.shape,b)
```

```python
# On définit notre modèle
def perceptron(X,W,b):
    # On applique nos formules vectorisées
    Z=X.dot(W)+b
    #print(Z.max())
    A=1/(1+np.exp(-Z))
    return A
```

```python
A=perceptron(Xr,W,b)
print(A)
```

```python
print(y)
```

```python
# Fonction de prédiction par seuil
def predict(X,W,b):
    A=perceptron(X,W,b)
    #print(A)
    return A>0.5
```

### Fonction d'erreur

```python
# pour la minimisation de l'erreur quadratique
def errQ(A,y):
    m=len(y)
    return np.sum((A-y)**2)/m
```

```python
 print(errQ(A,y))
```

### Calcul du vecteur gradient

```python
# On détermine le gradient de l'erreur quadratique
def gradient(A,X,y):
    m=len(y)
    db=np.sum((A-y)*A*(1-A)) / m
    dW=np.dot(X.T,(A-y)*A*(1-A)) / m
    return (dW,db)
```

```python
(dW,db)=gradient(A,Xr,y)
print(dW)
```

### Descente de gradient

```python
# on met à jour les paramètres par descente de gradient
def mAj(dW,db,W,b,alpha):
    W=W-alpha*dW
    b=b-alpha*db
    return (W,b)
```

### Propagation arrière par descente de gradient

```python
def learnRN(X,y,alpha=0.1, iter=100): 
    # initialisation
    n=X.shape[1]
    W,b = initParam(n)
    LerQ=[]
    for i in range(iter):# on fixe ici le nb d'itérations pour la descente de gradient
        # Forward propagation : On calcule Z puis A
        A = perceptron(X,W,b)
        # puis l'erreur quadratique
        err=errQ(A,y)
        LerQ.append(err)
        # Backward propagation : On calcule les gradients
        dW,db = gradient(A,X,y)
        # et on met à jour les paramètres
        W,b =  mAj(dW,db,W,b,alpha)
        
    plt.plot (LerQ)  # On trace la courbe de descente des couts 
    plt.show()
    return W,b
```

```python
learnRN(Xr,y)
```

Observer les valeurs de Z à chaque étape et comprendre le problème! On a effectivement de très grandes valeurs (positives mais aussi négatives) pour Z. Donc pour A nous avons des alternances de 0 et de 1, donc des erreurs de 0 ou 1 peu nuancées.

C'est dû au fait que les valeurs des pixels sont dans [0,256]. On ajoute donc 4096 valeurs dans [0.255] ce qui est rédhibitoire pour une exponentielle. On va donc d'abord normaliser X dans [0,1] Cette normalisation est très souvent nécessaire en deep learning, surtout quand les variables ont des ordres de grandeur très différents.

On effectue par exemple une normalisation  : x=(x-xmin)/(xmax-xmin)=(X-0)/255=X/255 

```python
Xn=Xr/255
Xtn=Xt/255 # On fait bien sûr pareil sur les données de test
print(Xn.shape,Xn.max(),Xn.min())
```

Ca continue de descendre, mais il est possible que ce soit de l'overfitting.
Il est temps de calculer les performances prédictives du modèle : 

```python
# Quelle est alors la performance du modèle?
from sklearn.metrics import accuracy_score
```

```python
def learnRN(X,y,alpha=0.1, iter=100): 
    # initialisation
    n=X.shape[1]
    W,b = initParam(n)
    LerQ=[]
    Acc=[]
    for i in range(iter):# on fixe ici le nb d'itérations pour la descente de gradient
        # Forward propagation : On calcule Z puis A
        A = perceptron(X,W,b)
        # puis l'erreur quadratique
        err=errQ(A,y)
        LerQ.append(err)
        yPred=predict(X,W,b)
        Acc.append(accuracy_score(y,yPred))
        # Backward propagation : On calcule les gradients
        dW,db = gradient(A,X,y)
        # et on met à jour les paramètres
        W,b =  mAj(dW,db,W,b,alpha)
        
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot (LerQ)  # On trace la courbe de descente des couts 
    plt.subplot(1,2,2)
    plt.plot(Acc)
    plt.show()
    return W,b
```

```python
W,b = learnRN(Xn,y,alpha=0.1,iter=600)
```

Ca continue d'augmentert, c'est très encourageant! (même si on ne dépasse pas beaucoup les 50%)
Il est grand temps de regarder comment le modèle se comporte sur les données de test!

```python
# on ajoute le traitement des données de test
def learnRN(X,y,Xt,yt,alpha=0.1, iter=100): 
    # initialisation
    n=X.shape[1]
    W,b = initParam(n)
    LerQ=[]# liste des coûts
    Acc=[] # liste des performances du modèle
    # et sur les données de test
    LerQT=[]# liste des coûts
    AccT=[]
    for i in range(iter):# on fixe ici le nb d'itérations pour la descente de gradient
        # On calcule Z puis A
        A = perceptron(X,W,b)
        
        if i%10==0:# on ne calcule les scores qu'une fois sur 10
            # puis la fonction de cout
            err=errQ(A,y)
            LerQ.append(err)
            yPred=predict(X,W,b)
            Acc.append(accuracy_score(y,yPred))
            
            # et sur les données de test
            At = perceptron(Xt,W,b)
            errt=errQ(At,yt)
            LerQT.append(errt)
            yPredT=predict(Xt,W,b)
            AccT.append(accuracy_score(yt,yPredT))
        # On calcule les gradients
        dW,db = gradient(A,X,y)
        # et on met à jour les paramètres
        W,b =  mAj(dW,db,W,b,alpha)
    
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot (LerQ,label='Données d''entraînement')  # On trace la courbe de descente des couts 
    plt.plot (LerQT,label='Données de test')
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(Acc,label='Données d''entraînement')
    plt.plot(AccT,label='Données de test')
    plt.legend()
    plt.show()
    return W,b
```

```python
W,b = learnRN(Xn,y,Xtn,yt,alpha=0.1,iter=10000)
```

C'était bel et bien de l'overfitting. On plafonne à un petit 57%...

## 2)  On recommence tout avec l'optimisation de la vraissemblance...

On calcule les nouvelles fonctions de performance et les nouveaux gradients

```python
#pour la maximisation de la vraisemblance
def logloss(A,y):
    eps=1e-15
    m=len(y)
    return -(np.sum(y*np.log(A+eps)+(1-y)*np.log(1-A+eps)))/m
```

```python
def learnRN(X,y,alpha=0.1, iter=100): 
    # initialisation
    n=X.shape[1]
    W,b = initParam(n)
    LerQ=[]
    for i in range(iter):# on fixe ici le nb d'itérations pour la descente de gradient
        # Forward propagation : On calcule Z puis A
        A = perceptron(X,W,b)
        # puis l'erreur quadratique
        err=logloss(A,y)
        LerLog.append(err)
        # Backward propagation : On calcule les gradients
        dW,db = gradient(A,X,y)
        # et on met à jour les paramètres
        W,b =  mAj(dW,db,W,b,alpha)
        
    plt.plot (LerQ)  # On trace la courbe de descente des couts 
    plt.show()
    return W,b
```
