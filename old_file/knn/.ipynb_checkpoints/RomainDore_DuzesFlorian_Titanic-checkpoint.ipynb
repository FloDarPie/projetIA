### 1) Chargement et préparation des données

```python
#On commence par importer nos bibliothèques habituelles
import sklearn
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
titanic=sb.load_dataset('titanic') # on importe le data set depuis Seaborn
titanic.head()
```

```python
# Modifier les données de la manière suivante :
# on ne garde que ces 4 caractéristiques : 'survived','pclass','sex','age'

#mini_titanic = titanic.drop(['sibsp', 'parch', 'fare', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone'], axis = 1)
miniTitanic = {
    'survived' : titanic['survived']
    ,'pclass' : titanic['pclass']
    ,'sex' : titanic['sex']
    ,'age' : titanic['age']
}

titanic = pd.DataFrame(miniTitanic)
```

```python
# on élimine les données incomplètes
titanic = titanic.dropna()
```

```python
# on remplace male par 0 et female par 1
titanic['sex'].replace({'female': 1, 'male': 0}, inplace = True)
```

```python
#affichage
titanic.head()
```

```python
# la colonne survided sera notre y et class, sexe et âges nos caractéristiques X

Y = titanic['survived']


X = titanic.drop(['survived'],axis = 1)
```

```python
titanic.groupby(['pclass']).mean()
```

## Recherche de l'IA

```python
from sklearn.neighbors import KNeighborsClassifier as knc
knn = knc(n_neighbors = 1)
```

```python
# à l'aide de train_test_split,
# partager les données en un jeu d'apprentissage et un jeu de test
from sklearn.model_selection import train_test_split as ttc
Xa, Xt, Ya, Yt = ttc(X, Y, random_state=0)
```

```python
# à l'aide de fit, associer le jeu de données d'apprentissage à votre knn
knn.fit(Xa, Ya)

#https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
```

```python
# Tester votre classificateur knn sur les données de test à l'aide de predict
# et comparer les résultats obtenus avec les vraies valeurs
prediction = knn.predict(Xt)
```

```python
#Quel est le score de prédiction de votre knn? 
score = knn.score(Xt,Yt, sample_weight = None)
print(score)
```

```python
def recuperation_du_meilleur():
    score = []

    for nombre_voisin in range(1,30):
        knn = knc(n_neighbors = nombre_voisin)
        Xa, Xt, Ya, Yt = ttc(X, Y, random_state=0)
        knn.fit(Xa, Ya)
        score.append(knn.score(Xt,Yt, sample_weight = None))
    plt.plot(score)

    maximum = max(score)

    print('meilleurs parametres :', score.index(maximum))
    print('meilleurs score :', maximum)
```

```python
recuperation_du_meilleur()
```

```python
# Effectuer une validation croisée à 5 répétitions, soit à la main, 
# soitl'aide de cross_val_score

from sklearn.model_selection import cross_val_score as cvs
print(cvs(knn, X, Y, cv = 10))
```

```python
def recuperation_du_meilleur_avec_validation(nbValidation):
    score = []

    for nombre_voisin in range(1,30):
        knn = knc(n_neighbors = nombre_voisin)
        listeScore = cvs(knn, X, Y, cv = nbValidation)
        
        moyenne = 0
        for resultat in listeScore:
            moyenne += resultat
        
        score.append(moyenne / nbValidation)
    plt.plot(score)

    maximum = max(score)

    print('meilleurs parametres :', score.index(maximum))
    print('meilleurs score :', maximum)
```

```python
recuperation_du_meilleur_avec_validation(20)
```

```python
def survie(nombre_voisin,X, Y,  classe,sex,age):
    knn = knc(n_neighbors = nombre_voisin)
    
    Xa, Xt, Ya, Yt = ttc(X, Y)
    knn.fit(Xa, Ya)
    
    indv = {
    'pclass' : [classe]
    ,'sex' : [sex]
    ,'age' : [age]
    }
    
    individu = pd.DataFrame(indv) 
    
    prediction = knn.predict(individu)
    
    if sex:
        print("Test pour une femme de %s ans en classe %s" % (age,classe))
    else :
        print("Test pour un homme de %s ans en classe %s" % (age,classe))

    if prediction[0] == 1:
        return "La personne survit"
    else :
        return "La personne ne survit pas"
```

```python
#Le nombre de voisin est mis sur 1 car il s'agit du meilleur resultat donnes par l'etude precedente

# --- Paramètres ---
nombre_voisins = 1
classe = 2
sex = 0
age = 60
# ------------------
    
print(survie(nombre_voisins,X,Y, classe,sex,age))
```

# Utilisation de GridSearchCV

```python
from sklearn.model_selection import GridSearchCV as GSCV
from sklearn.neighbors import KNeighborsClassifier as KNC

parameters = {
    'n_neighbors' : list(range(1, 30)),
    'weights' : ['uniform', 'distance'],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
    #'metrics_params' : ['minkowski']
}

#euclidean','manhattan','minkowski','wminkowski','seuclidean','mahalanobis'
neigh = KNC()

#print(estimator.get_params().keys())

clf = GSCV(neigh, parameters, verbose=2, return_train_score=True)
clf.fit(X, Y)

print('meilleurs parametres :', clf.best_params_)
print('meilleurs score :', clf.best_score_)
```

```python
res = pd.DataFrame(clf.cv_results_)
print(res.head())
```

```python
#meilleurs parametres : {'algorithm': 'auto',
#                       'n_neighbors': 26,
#                       'weights': 'distance'}


# on enleve les lignes qui nous interessent pas
res = res.loc[res["param_weights"] == 'distance']
res = res.loc[res["param_algorithm"] == 'auto']

miniRes = {
    'test' : res['mean_test_score']
    ,'train' : res['mean_train_score']
    ,'n_neighbours' : res['param_n_neighbors']
    #,'param_weights' : res['param_weights']
    #,'param_algorithm' : res['param_algorithm']
}

resultat = pd.DataFrame(miniRes)
```

```python
resultat.head()
```

```python
k = res['param_n_neighbors']
test_score = res['mean_test_score']
train_score = res['mean_train_score']


df_score = pd.DataFrame(dict(k=k, test=test_score, train=train_score))
ax = df_score.plot(x='k', y='train', figsize=(6, 4))
df_score.plot(x='k', y='test', ax=ax, grid=True)


ax.set_title("Evolution de la performance sur\nles bases d'apprentissage et de test" +
             "\nen fonction du nombre de voisins")
             
```

```python
def GSV_Knn(parameters, titanic, Knn_function):
    
    #reduction database
    miniTitanic = {
    'survived' : titanic['survived']
    ,'pclass' : titanic['pclass']
    ,'sex' : titanic['sex']
    ,'age' : titanic['age']
    }

    titanic = pd.DataFrame(miniTitanic)
    
    #construction de notre jeu de test
    Y = titanic['survived']
    X = titanic.drop(['survived'],axis = 1)

    #choix du Knn_
    neigh = Knn_function
    
    #recherche de la meilleure solution
    clf = GSCV(neigh, parameters, verbose=2, return_train_score=True)
    clf.fit(X, Y)
    
    #recuperation des meilleurs elements
    weight = clf.best_params_['weights']
    algo = clf.best_params_['algorithm']
    
    #construction du dataset des resultats
    res = pd.DataFrame(clf.cv_results_)
    
    #nettoyage des resultats
    res = res.loc[res["param_weights"] == weight]
    res = res.loc[res["param_algorithm"] == algo]
    miniRes = {
    'test' : res['mean_test_score']
    ,'train' : res['mean_train_score']
    ,'n_neighbours' : res['param_n_neighbors']
    }

    resultat = pd.DataFrame(miniRes)
    
    
    #affichage des resultats
    print('meilleurs parametres :', clf.best_params_)
    print('meilleurs score :', clf.best_score_)
    
    #graphique
    k = resultat['n_neighbours']
    test_score = resultat['test']
    train_score = resultat['train']


    df_score = pd.DataFrame(dict(k=k, test=test_score, train=train_score))
    ax = df_score.plot(x='k', y='train', figsize=(6, 4))
    df_score.plot(x='k', y='test', ax=ax, grid=True)


    ax.set_title("Evolution de la performance sur\nles bases d'apprentissage et de test" +
                 "\nen fonction du nombre de voisins")
```

```python
def GSV_Knn_resultat(parameters, titanic, Knn_function):
    
    #reduction database
    miniTitanic = {
    'survived' : titanic['survived']
    ,'pclass' : titanic['pclass']
    ,'sex' : titanic['sex']
    ,'age' : titanic['age']
    }

    titanic = pd.DataFrame(miniTitanic)
    
    #construction de notre jeu de test
    Y = titanic['survived']
    X = titanic.drop(['survived'],axis = 1)

    clf = GSCV(Knn_function, parameters)
    clf.fit(X, Y)

    print('meilleurs parametres :', clf.best_params_)
    print('meilleurs score :', clf.best_score_)
```

```python
parameters = {
    'n_neighbors' : list(range(1, 50)),
    'weights' : ['uniform', 'distance'],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
}

Knn_function = KNC()
```

```python
GSV_Knn(parameters, titanic, Knn_function)
```

```python
from sklearn.neighbors import RadiusNeighborsRegressor as RNR

Knn_function = RNR()
```

```python
parameters = {
    'radius' : list( i + 0.5 for i in range(0,20)),
    'weights' : ['uniform', 'distance'],
    'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
}
```

```python
GSV_Knn_resultat(parameters, titanic, Knn_function)
```

```python
def survie(titanic, age, sex, classe):
    from sklearn.neighbors import KNeighborsClassifier as KNC
    #reduction database
    miniTitanic = {
    'survived' : titanic['survived']
    ,'pclass' : titanic['pclass']
    ,'sex' : titanic['sex']
    ,'age' : titanic['age']
    }

    titanic = pd.DataFrame(miniTitanic)
    
    #construction de notre jeu de test
    Y = titanic['survived']
    X = titanic.drop(['survived'],axis = 1)
    
    
    #construction individu
    indv = {
    'pclass' : [classe]
    ,'sex' : [sex]
    ,'age' : [age]
    }
    individu = pd.DataFrame(indv)
    
    
    clf = GSCV(KNC(), {'algorithm': ['auto'], 'n_neighbors': [26], 'weights': ['distance']})
    clf.fit(X, Y)
    #clf.set_params(n_neighbors=26,weight='distance')

    print('Resultat :', clf.score(individu))
```

```python
survie(titanic, 20,0,2)
```
